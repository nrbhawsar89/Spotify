### Preprocessing Decisions
--1) type column can be dropped

2) Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.

**3) The key the track is in. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1.

**4) instrumentalness (correlation with long duration track) Predicts whether a track contains no vocals. Ooh and aah sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly vocal. The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0.

**5) A value of 0.0 is least danceable and 1.0 is most danceable.

**6) Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity.

**7) loudness: Values typically range between -60 and 0 db.

**8) Speechiness detects the presence of spoken words in a track. 

**9) acousticness: Acoustic music is music that solely or primarily uses instruments that produce sound through acoustic means, as opposed to electric or electronic means.

**10) liveness Detects the presence of an audience in the recording.

**11) valence: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).

--12) Convert duration_ms to duration_min and drop duration_ms

**13) A time signature (also known as meter signature, metre signature, and measure signature) is a convention in Western music notation that specifies how many note values of a particular type are contained in each measure (bar). (0/0 time signature can be used to denote free time)

--14) split the df with id,uri,href,analysis_url

```python
features_df['time_signature'].unique()
```

```python
features_df.drop(columns=['type'],inplace=True)

features_df['duration_min']= np.round((features_df['duration_ms']/1000)/60)

features_df.drop(columns=['duration_ms'], inplace=True)

track_df = features_df[['id','uri','track_href','analysis_url']]
track_df.head()

features_df.drop(columns=['uri','track_href','analysis_url'], inplace=True)

features_df.rename(columns={
    'danceability':'Danceability',
    'energy':'Energy',
    'key':'Key',
    'loudness':'Loudness',
    'mode':'Mode',
    'speechiness':'Lyrical',
    'acousticness':'Acousticness',
    'instrumentalness':'Instrumental',
    'liveness':'Live_Audience',
    'valence':'Positiveness',
    'tempo':'Tempo',
    'id':'Track_ID',
    'time_signature':'TSignature',
    'duration_min':'Duration(min)'
}, inplace=True)
```

```python
# >=0 and <=3: {'minute_range':3, 'minute_range_text':'0-3 min'} 
# >3 and <=5: {'minute_range':5, 'minute_range_text':'3-5 min'} 
# >5 and <=7: {'minute_range':7, 'minute_range_text':'5-7 min'},
# >7 and <=8: {'minute_range':8, 'minute_range_text':'7-8 min'},
# >8 and <=10:{'minute_range':10, 'minute_range_text':'8-10 min'},
# >10 and <=15:{'minute_range':15, 'minute_range_text':'10-15 min'},
# >15 and <=30:{'minute_range':30, 'minute_range_text':'30 min'},
# >30 and <=45:{'minute_range':45, 'minute_range_text':'45 min'},
# >45 and <=60:{'minute_range':60, 'minute_range_text':'1 hr'},
# >60 and <=180:{'minute_range':180, 'minute_range_text':'2-3 hr'},
# >180 and <=360:{'minute_range':360, 'minute_range_text':'3-6 hr'},
# >360 and <=720:{'minute_range':720, 'minute_range_text':'12 hr'},
# >720 :{'minute_range':1000, 'minute_range_text':'Outlier'},

def get_minute_range(minutes):
    if minutes >= 0 and minutes <= 3:
        return 3, '0-3 min'
    elif minutes > 3 and minutes <= 5:
        return 5, '3-5 min'
    elif minutes > 5 and minutes <= 7:
        return 7, '5-7 min'
    elif minutes > 7 and minutes <= 8:
        return 8, '7-8 min'
    elif minutes > 8 and minutes <= 10:
        return 10, '8-10 min'
    elif minutes > 10 and minutes <= 15:
        return 15, '10-15 min'
    elif minutes > 15 and minutes <= 30:
        return 30, '30 min'
    elif minutes > 30 and minutes <= 45:
        return 45, '45 min'
    elif minutes > 45 and minutes <= 60:
        return 60, '1 hr'
    elif minutes > 60 and minutes <= 180:
        return 180, '2-3 hr'
    elif minutes > 180 and minutes <= 360:
        return 360, '3-6 hr'
    elif minutes > 360 and minutes <= 720:
        return 720, '12 hr'
    else:
        return 1000, 'Outlier'
```

```python
features_df['Minute_range'], features_df['Minute_range_text'] = zip(*features_df['Duration(min)'].apply(get_minute_range))
```

```python
# calculate the correlation matrix on the numeric columns
corr = features_df.select_dtypes('number').corr()

# plot the heatmap
sns.heatmap(corr)
```

```python

minute_range_agg_df = features_df.groupby(['Minute_range','Minute_range_text']).agg({
    'Track_ID':'nunique',
    'Danceability': 'mean',
    'Energy':'mean',
    'Key':'mean',
    'Loudness':'mean',
    'Lyrical':'mean',
    'Acousticness':'mean',
    'Instrumental':'mean',
    'Live_Audience':'mean'
}).reset_index()
```
